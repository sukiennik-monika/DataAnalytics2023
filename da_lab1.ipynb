```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from scipy.stats import norm, beta
```

##### ex. 2

```python
arr_uniform_distrib = np.random.uniform(0, 1, 1000)
arr_normal_distrib_stand_dev_1 = np.random.normal(0, 1, 1000)
arr_normal_distrib_stand_dev_10 = np.random.normal(0, 10, 1000)
arr_beta_distrib_ab_11 = np.random.beta(1,1,1000)
arr_beta_distrib_ab_22 = np.random.beta(2,2,1000)
arr_beta_distrib_ab_39 = np.random.beta(3,9,1000)
arr_2d_normal_distrib_cov_1001 = np.random.multivariate_normal([0,0], [[1,0],[0,1]], 10000)
arr_2d_normal_distrib_cov_2112 = np.random.multivariate_normal([0,0], [[2,1],[1,2]], 10000)
```

##### ex. 3

```python
plt.figure(1)
hist_arr_uniform_distrib, bins, ignored = plt.hist(arr_uniform_distrib, 20, density=True)
plt.plot(bins, np.ones_like(bins),color='r')
plt.title(f'Comparison of Uniform Distribution and histogram for array with this distribution')
plt.ylabel(f'Density')
plt.xlabel(f'Values')
plt.show()
```

```python
plt.figure(2)
hist_arr_normal_distrib_stand_dev_1 = plt.hist(arr_normal_distrib_stand_dev_1, 20, density=True)
x_min, x_max = plt.xlim()
x_values = np.linspace(x_min,x_max, 1000)
p_values = norm.pdf(x_values, 0, 1)
plt.plot(x_values, p_values, 'r', linewidth=2)
plt.title(f'Comparison of Normal Distribution and histogram for array with this distribution and standard devation equals 1')
plt.ylabel(f'Density')
plt.xlabel(f'Values')
plt.show()
```

```python
plt.figure(3)
hist_arr_normal_distrib_stand_dev_10 = plt.hist(arr_normal_distrib_stand_dev_10, 20, density=True)
x_min, x_max = plt.xlim()
x_values = np.linspace(x_min,x_max, 1000)
p_values = norm.pdf(x_values, 0, 10)
plt.plot(x_values, p_values, 'r', linewidth=2)
plt.title(f'Comparison of Normal Distribution and histogram for array with this distribution and standard devation equals 10')
plt.ylabel(f'Density')
plt.xlabel(f'Values')
plt.show()
```

```python
plt.figure(4)
hist_arr_beta_distrib_ab_11 = plt.hist(arr_beta_distrib_ab_11, 20, density=True)
x_values = np.linspace(beta.ppf(0, 1, 1), beta.ppf(1, 1, 1), 1000)
plt.plot(x_values, beta.pdf(x_values, 1, 1),'r', lw=2)
plt.title(f'Comparison of Beta Distribution and histogram for array with this distribution and params (a,b) = (1,1)')
plt.ylabel(f'Density')
plt.xlabel(f'Values')
plt.show()
```

```python
plt.figure(5)
hist_arr_beta_distrib_ab_22 = plt.hist(arr_beta_distrib_ab_22, 20, density=True)
x_values = np.linspace(beta.ppf(0, 2, 2), beta.ppf(1, 2, 2), 1000)
plt.plot(x_values, beta.pdf(x_values, 2, 2),'r', lw=2)
plt.title(f'Comparison of Beta Distribution and histogram for array with this distribution and params (a,b) = (2,2)')
plt.ylabel(f'Density')
plt.xlabel(f'Values')
plt.show()
```

```python
plt.figure(6)
hist_arr_beta_distrib_ab_39 = plt.hist(arr_beta_distrib_ab_39, 20, density=True)
x_values = np.linspace(beta.ppf(0, 3, 9), beta.ppf(1, 3, 9), 1000)
plt.plot(x_values, beta.pdf(x_values, 3, 9),'r', lw=2)
plt.title(f'Comparison of Beta Distribution and histogram for array with this distribution and params (a,b) = (3,9)')
plt.ylabel(f'Density')
plt.xlabel(f'Values')
plt.show()
```

##### ex. 5

```python
alfa = np.random.normal(2, 0.1, 1000)
beta = np.random.normal(0.2, 0.2, 1000)
x = np.random.normal(0, 3, 1000)
sigma = np.abs(np.random.normal(0, 1, 1000))
my = alfa + beta * x
y = np.random.normal(my, sigma, 1000)
```

##### ex. 6

```python
plt.figure(11)
plt.scatter(x, y)
x_plot = np.linspace(-6, 6, 100)
for i in range(1000):
    y_plot = alfa[i] + beta[i] * x_plot
    plt.plot(x_plot, y_plot)
plt.xlim(-6,6)
plt.title(f'Scatter plot of data and lines for y = alfa + beta * x')
plt.show()
```
